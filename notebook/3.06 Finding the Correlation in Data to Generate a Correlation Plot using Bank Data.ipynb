```python
import pandas as pd
file_url = r'https://raw.githubusercontent.com/sedeba19/Chapter-3/main/data/bank-full.csv'
bankData = pd.read_csv(file_url, sep=";")
```

```python
from pandas import set_option
```

```python
bankNumeric = bankData[['age',
                        'balance',
                        'day',
                        'duration',
                        'campaign',
                        'pdays',
                        'previous']]
```

```python
# Now, use the .corr() function to find the correlation matrix for the dataset:
correlations = bankNumeric.corr(method='pearson')
print(correlations)
```

```python
# Plot the data
import matplotlib.pyplot as plt
corFig = plt.figure()
figAxis = corFig.add_subplot(111)
corAx = figAxis.matshow(correlations, vmin=-1, vmax=1)
corFig.colorbar(corAx)
plt.show()
```
