```python
import pandas as pd
import numpy as np
```

```python
file_url = r'https://raw.githubusercontent.com/sedeba19/Chapter-3/main/data/bank-full.csv'
```

```python
# Reading the banking data
bankData = pd.read_csv(file_url,sep=";")
```

```python
bankData.head()
```

```python
# Normalizing data
from sklearn import preprocessing
x = bankData[['balance']].values.astype(float)

minmaxScaler = preprocessing.MinMaxScaler()
bankData['balanceTran'] = minmaxScaler.fit_transform(x)
bankData
```

```python
# Adding a small numerical constant to eliminate 0 values
bankData['balanceTran'] = bankData['balanceTran'] + 0.00001
bankData
```

```python
# Let us transform values for loan data
bankData['loanTran'] = 1
# Giving a weight of 5 if there is no loan
bankData.loc[bankData['loan'] == 'no', 'loanTran'] = 5
bankData.head()
```

```python
# Let us transform values for Housing data
bankData['houseTran'] = 5
bankData.loc[bankData['housing'] == 'no', 'houseTran'] = 1
bankData.head()
```

```python
# Let us now create the new variable which is a product of all these
bankData['assetIndex'] = bankData['balanceTran']* bankData['loanTran']* bankData['houseTran']
bankData.head()
```

```python
# Finding the quantile
np.quantile(bankData['assetIndex'],[0.25,0.5,0.75])
```

```python
# Create quantiles from the `assetindex` data
bankData['assetClass'] = 'Quant1'
bankData.loc[(bankData['assetIndex'] > 0.38) & (bankData['assetIndex'] < 0.57),
              'assetClass'] = 'Quant2'
bankData.loc[(bankData['assetIndex'] > 0.57) & (bankData['assetIndex'] < 1.9), 
              'assetClass'] = 'Quant3'
bankData.loc[bankData['assetIndex'] > 1.9,
             'assetClass'] = 'Quant4'
#bankData.assetClass[bankData['assetIndex'] > 1.9] = 'Quant4'
bankData.head()
```

```python
# Calculating total of each asset class
assetTot = bankData.groupby('assetClass')['y'].agg(assetTot='count').reset_index()
# Calculating the category wise counts
assetProp = bankData.groupby(['assetClass', 'y'])['y'].agg(assetCat='count').reset_index()
```

```python
assetTot
```

```python
assetProp
```

```python
# Merging both the data frames
assetComb = pd.merge(assetProp, assetTot, on = ['assetClass'])
assetComb['catProp'] = (assetComb.assetCat / assetComb.assetTot)*100
assetComb
```

```python
# Categorical variables, removing loan and housing
bankCat1 = pd.get_dummies(bankData[['job',
                                    'marital',
                                    'education',
                                    'default',
                                    'contact',
                                    'month',
                                    'poutcome']])
bankCat1.shape
```

```python
# Numerical variables
bankNum1 = bankData[['age','day','duration','campaign',\
                     'pdays','previous','assetIndex']]
bankNum1.head()
```

```python
bankNum1.dtypes
```

```python
# Normalize the numerical variables
from sklearn.preprocessing import MinMaxScaler
minmaxScaler = preprocessing.MinMaxScaler()

# Transforming the numerical variables
ageT1 = bankNum1[['age']].values.astype(float)
dayT1 = bankNum1[['day']].values.astype(float)
durT1 = bankNum1[['duration']].values.astype(float)

# Creating numerical normalized variables
bankNum1['ageTran'] = minmaxScaler.fit_transform(ageT1)
bankNum1['dayTran'] = minmaxScaler.fit_transform(dayT1)
bankNum1['durTran'] = minmaxScaler.fit_transform(durT1)

# Create a new variable for numerical normalized variables
bankNum2 = bankNum1[['ageTran','dayTran','durTran',\
                     'campaign','pdays','previous',\
                     'assetIndex']]
bankNum2.head()
```

```python
# Merging numerical and categorical variables
bankComb = pd.concat([bankNum2, bankCat1], axis = 1)
bankComb.head()
```

```python
# Dependent variable
bankDep = bankData['y']
bankDep.head()
```

```python
# Splitting the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(bankComb, 
                                                    bankDep,
                                                    test_size = 0.3,
                                                    random_state = 123)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Building the logistic regression model
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression(random_state=123)
logReg.fit(X_train, y_train)
```

```python
# Predicting the output classes
y_pred = logReg.predict(X_test)
y_pred
```

```python
# Accuracy of the model
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
